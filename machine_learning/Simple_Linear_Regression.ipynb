```python
# ## Question 1: Your First Model in Python üêç
# A student wants to see if there's a relationship between the number of hours they study and the score they get on a quiz.

# Dataset:

# Python

# import pandas as pd

# data1 = {
#     'Hours_Studied': [1, 2, 3, 4, 5],
#     'Quiz_Score': [65, 70, 80, 85, 95]
# }
# df1 = pd.DataFrame(data1)
# Your Tasks:

# Prepare the data: Create the pandas DataFrame as shown above. Define your feature X (Hours_Studied) and target Y (Quiz_Score). Remember that X needs to be a 2D array for scikit-learn, so you'll need to reshape it (e.g., X = df1[['Hours_Studied']]).

# Build and train the model: Import LinearRegression from sklearn.linear_model, create an instance of the model, and fit it to your data.

# Find the equation: Print the model's coefficient (Œ≤ 
# 1
# ‚Äã
#  ) and intercept (Œ≤ 
# 0
# ‚Äã
#  ).

# Make a prediction: Use the trained model to predict the quiz score for a student who studied for 3.5 hours.

# Evaluate the model: Import r2_score from sklearn.metrics. Calculate the R-squared value for your model and print it.


import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression
import numpy as np

data1 = {
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Quiz_Score': [65, 70, 80, 85, 95]
}
df = pd.DataFrame(data1)

# split data into separate arrays, 2d for SciKit learn
x = df[['Hours_Studied']] #independent variable, X
y = df['Quiz_Score']      #label, something that we need to predict


model = LinearRegression()
model.fit(x,y)

slope = model.coef_[0]
y_intercept = model.intercept_

y_predict = model.predict(x)

r_squared = r2_score(y,y_predict)

print("b1 = ", slope)
print("b0 = ", y_intercept)
print("prediction for 3.5 =", model.predict([[3.5]]))
print("r_square =", r_squared)

plt.scatter(x,y, color = 'b')
plt.plot(x,y_predict, color = 'r')
plt.show()
```

```python
# ## Question 2: Visualization and Error Metrics üìä
# A small coffee shop is tracking its daily advertising spend against the number of coffees sold.

# Dataset:

# Python

# import pandas as pd

# data2 = {
#     'Ad_Spend': [10, 15, 20, 25, 30, 35, 40, 50],
#     'Coffees_Sold': [95, 110, 118, 140, 145, 160, 155, 180]
# }
# df2 = pd.DataFrame(data2)
# Your Tasks:

# Model the relationship: Create a DataFrame, define X and Y, and train a simple linear regression model.

# Visualize the results:

# Create a scatter plot of the actual data points (Ad Spend vs. Coffees Sold) using matplotlib.

# Use your model to predict Y values for the existing X values.

# Plot the regression line over your scatter plot to see how well it fits.

# Interpret the parameters: In a comment or markdown cell, explain what the learned slope (Œ≤ 
# 1
# ‚Äã
#  ) and intercept (Œ≤ 
# 0
# ‚Äã
#  ) mean in the context of ad spending and coffee sales.

# Calculate the error: Import mean_squared_error from sklearn.metrics. Calculate the MSE for your model.

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, root_mean_squared_error
import matplotlib.pyplot as plt

data2 = {
    'Ad_Spend': [10, 15, 20, 25, 30, 35, 40, 50],
    'Coffees_Sold': [95, 110, 118, 140, 145, 160, 155, 180]
}
df2 = pd.DataFrame(data2)
x_actual = df2[['Ad_Spend']]
y_actual = df2['Coffees_Sold']

model = LinearRegression()
model.fit(x_actual,y_actual)

y_predict = model.predict(x_actual)


plt.scatter(x_actual,y_actual, color = 'blue')
plt.xlabel("AD Spend")
plt.ylabel("Coffees Sold")
plt.plot(x_actual,y_predict,color = 'red')



print("intercept = ", model.intercept_)
print("coefficient of x =", model.coef_[0])
#from the intercept value i can learn that the dependent variable "Coffees sold" is impacted by 2.07556 times Ad Spend, in other terms, for every 1 dollar increase in Advertisement expenditure we can expect 2.07 coffees sold approximately . The intercept tells me that if there is no Ad Spend then i can expect about 79.496 coffees sold daily.

print("mean squared error", mean_squared_error(y_actual,y_predict))
```

```python
# Question 3: Negative Correlation and RMSE üìâ
# A used car dealership wants to model the relationship between the age of a car and its resale price.

# Dataset:

# Python

# import pandas as pd

# data3 = {
#     'Car_Age_Years': [1, 2, 3, 4, 5, 6, 7, 8],
#     'Price_in_thousands': [28, 25, 21, 22, 17, 15, 12, 10]
# }
# df3 = pd.DataFrame(data3)
# Your Tasks:

# Build the model: Following the now-familiar steps, train a linear regression model to predict price based on age.

# Calculate RMSE:

# First, calculate the Mean Squared Error (MSE).

# Then, calculate the Root Mean Squared Error (RMSE) by taking the square root of the MSE (you can use numpy.sqrt()).

# Interpret RMSE: In a comment, explain what the RMSE value means. (Hint: It's in the same units as your target variable, 'Price in thousands').

# Assess the fit: Calculate the R 
# 2
#   value. How well does the car's age explain the variance in its price according to your model?

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, root_mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

#setting up the data
data3 = {
    'Car_Age_Years': [1, 2, 3, 4, 5, 6, 7, 8],
    'Price_in_thousands': [28, 25, 21, 22, 17, 15, 12, 10]
}
df = pd.DataFrame(data3)

#setting up the arrays
x = df[['Car_Age_Years']]
y = df['Price_in_thousands']

#setting up the model
model = LinearRegression()
#model fitting
model.fit(x,y)
print("intercept = ", model.intercept_)
print("coefficient/slope = ", model.coef_)

#predicting using the model
y_predict = model.predict(x)

#plot to see the model line
plt.scatter(x,y, color = 'blue')
plt.plot(x,y_predict, color = 'red')
plt.xlabel("Car Age in Years")
plt.ylabel("Price in Thousands")
plt.show();

#questions answered
print("r square = ", r2_score(y, y_predict))
print("mean squared error = ", mean_squared_error(y, y_predict))
print("root mean squared error = ", root_mean_squared_error(y, y_predict))

#root mean square error is very low in value, while mean squared error is generally used to punish the high error observations severely the root mean squared error is considered to bring back the error back into the original units as the observation. in our case we can expect about 0.9290 units of error in prediction, or in simple terms we can say that the prediction about the price of a car could be off about 0.9290 thousand dollars for each prediction we make. 
#r square tells us it the model is a strong fit. also, looking at the graph i can easily see its a negative relation between car's age and its price in thousand. Terms, 95% of the variance in car's price can be explained by its age.
```

```python
# Question 5: The Outlier Challenge ü§Ø
# A real estate agent is modeling house prices based on their square footage. One data point looks suspicious.

# Dataset:

# Python

# import pandas as pd

# data5 = {
#     'Size_sq_ft': [1400, 1600, 1700, 1850, 2000, 2200, 2400, 1500, 2800, 3000],
#     'Price_in_thousands': [245, 312, 279, 308, 330, 390, 410, 550, 480, 500]
# }
# df5_with_outlier = pd.DataFrame(data5)
# Your Tasks:

# First Pass (With Outlier):

# Train a linear regression model on the complete df5_with_outlier DataFrame.

# Calculate and print the R-squared value and the RMSE.

# Optional but recommended: Create a scatter plot to visually confirm the outlier.

# Data Cleaning:

# Identify the outlier row (the house that is 1500 sq. ft. but costs $550k).

# Create a new DataFrame, df5_cleaned, by dropping that specific row.

# Second Pass (Without Outlier):

# Train a new linear regression model on the df5_cleaned data.

# Calculate and print the R-squared and RMSE for this new model.

# Compare and Conclude: In a markdown cell or a multi-line comment, compare the R-squared and RMSE from the first and second models. Explain briefly how that single data point impacted your model's performance and predictive accuracy.


import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, root_mean_squared_error
from sklearn.linear_model import LinearRegression
import numpy as np

data5 = {
    'Size_sq_ft': [1400, 1600, 1700, 1850, 2000, 2200, 2400, 1500, 2800, 3000],
    'Price_in_thousands': [245, 312, 279, 308, 330, 390, 410, 550, 480, 500]
}
df = pd.DataFrame(data5)


x = df[['Size_sq_ft']]
y = df['Price_in_thousands']

model = LinearRegression()
model.fit(x,y)
y_predict = model.predict(x)

print("intercept = ", model.intercept_)
print("slope/coefficient = ", model.coef_)
print("r_squared = ", r2_score(y,y_predict))
print("root mean squared error = ", root_mean_squared_error(y,y_predict))
print("mean of the distribution = ",np.average(x))

plt.scatter(x,y, color = 'blue')
plt.plot(x, y_predict, color = 'red')
plt.xlabel("Size in Square feet")
plt.ylabel("Price in thousands")
plt.show()
#clearly beacuse of the the outlier the mean, r_squared is very low - 0.301, RMSE Are showing up really high - 81.491.

df['Residuals'] = df['Price_in_thousands'] - y_predict

quartile_1 = df['Residuals'].quantile(0.25)
quartile_3 = df['Residuals'].quantile(0.75)
IQR = quartile_3 - quartile_1
low_spec = quartile_1 - 1.5*IQR
high_spec = quartile_3 + 1.5*IQR

df_cleaned = df[ (df['Residuals']>low_spec) & (df['Residuals']<high_spec) ]

x_clean = df_cleaned[['Size_sq_ft']]
y_clean = df_cleaned['Price_in_thousands']

model_2 = LinearRegression()
model_2.fit(x_clean,y_clean)
y_clean_predict = model_2.predict(x_clean)

print("intercept = ", model_2.intercept_)
print("slope/coefficient = ", model_2.coef_)
print("r_squared = ", r2_score(y_clean,y_clean_predict))
print("root mean squared error = ", root_mean_squared_error(y_clean,y_clean_predict))
print("mean of the distribution = ",np.average(x_clean))

plt.scatter(x_clean,y_clean, color = 'blue')
plt.plot(x_clean, y_clean_predict, color = 'red')
plt.xlabel("Size in Square feet")
plt.ylabel("Price in thousands")
plt.show()

#after clearing out the outlier the model fit is clearly better, the R2 was previously 0.312 but now it has increased to 0.969, also the RMSE has also reduced from 81.49 to 14.61. The outlier impacted the model a lot by taking increase the intercept and reducing the slope which led to increased error in predictions.
```
